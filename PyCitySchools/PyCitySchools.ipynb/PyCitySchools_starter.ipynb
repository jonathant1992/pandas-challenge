```python
import pandas as pd
```

```python
# Load school data
school_data = pd.read_csv("schools_complete.csv")
student_data = pd.read_csv("students_complete.csv")
```

```python
school_data.head()
student_data.head()
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
```

```python
total_schools = school_data["school_name"].nunique()
print(f"Total Schools: {total_schools}")

```

```python
total_students = student_data["Student ID"].count()
print(f"Total Students: {total_students}")
```

```python
total_budget = school_data["budget"].sum()
print(f"Total Budget: ${total_budget}")
```

```python
average_math_score = student_data["math_score"].mean()
print(f"Average Math Score: {average_math_score}")
```

```python
average_reading_score = student_data["reading_score"].mean()
print(f"Average Reading Score: {average_reading_score}")
```

```python
passing_math = student_data[student_data["math_score"] >= 70].count()["Student ID"]
percent_passing_math = (passing_math / total_students) * 100
print(f"Percentage Passing Math: {percent_passing_math}%")
```

```python
passing_reading = student_data[student_data["reading_score"] >= 70].count()["Student ID"]
percent_passing_reading = (passing_reading / total_students) * 100
print(f"Percentage Passing Reading: {percent_passing_reading}%")
```

```python
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading]
})

district_summary
```

```python

overall_passing = student_data[(student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70)]


percent_overall_passing = (overall_passing["Student ID"].count() / total_students) * 100

print(f"% Overall Passing: {percent_overall_passing}%")
```

```python
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing": [percent_overall_passing]
})

district_summary
```

```python

school_groups = school_data_complete.groupby("school_name")
```

```python
school_type = school_groups["type"].first()
```

```python
total_students_per_school = school_groups["Student ID"].count()
```

```python
total_school_budget = school_groups["budget"].first()
```

```python
per_student_budget = total_school_budget / total_students_per_school
```

```python
average_math_score = school_groups["math_score"].mean()
```

```python
average_reading_score = school_groups["reading_score"].mean()
```

```python
passing_math_per_school = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()
percent_passing_math = (passing_math_per_school / total_students_per_school) * 100
```

```python
passing_both_per_school = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name")["Student ID"].count()
percent_overall_passing = (passing_both_per_school / total_students_per_school) * 100
```

```python
passing_reading_per_school = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()
percent_passing_reading = (passing_reading_per_school / total_students_per_school) * 100
```

```python
school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_students_per_school,
    "Total School Budget": total_school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing": percent_overall_passing
})

# Display the school summary
school_summary
```

```python

top_schools = school_summary.sort_values("% Overall Passing", ascending=False)


top_schools.head(5)
```

```python

bottom_schools = school_summary.sort_values("% Overall Passing", ascending=True)


bottom_schools.head(5)
```
